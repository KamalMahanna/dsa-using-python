```python
def sortedSquares(nums: list[int]) -> list[int]:
    pos_idx=len(nums)
    # check where postive element starts
    for i in range(len(nums)):
        if nums[i]>=0:
            pos_idx = i
            break
    # if starts at begining return square of all elements
    if pos_idx==0:
        return [i**2 for i in nums]
    # if all are negetative, then reverse and square
    elif pos_idx==len(nums):
        return [nums[i]**2 for i in range(len(nums)-1,-1,-1)]
    else:
        new_nums=[]
        neg_idx=pos_idx-1
        while neg_idx>=0 and pos_idx<len(nums):
            if -nums[neg_idx]>nums[pos_idx]:
                new_nums.append(nums[pos_idx]**2)
                pos_idx+=1
            elif -nums[neg_idx]<nums[pos_idx]:
                new_nums.append(nums[neg_idx]**2)
                neg_idx-=1
            else:
                new_nums.append(nums[neg_idx]**2)
                new_nums.append(nums[neg_idx]**2)
                pos_idx+=1
                neg_idx-=1
        print(pos_idx,neg_idx)
        while neg_idx > -1:
            new_nums.append(nums[neg_idx]**2)
            neg_idx-=1
        while pos_idx<len(nums):
            new_nums.append(nums[pos_idx]**2)
            pos_idx+=1
    return new_nums
```

```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        counter=False
        left=0
        for i in range(1,len(nums)):
            if nums[i] == nums[left]:
                if not counter:
                    counter=True
                    left+=1
                    nums[left] = nums[i]
            else:
                left+=1
                nums[left] = nums[i]
                counter=False
        return left+1
```
